# Creating a 3-D Clustering Algorithm

```python
#Imported K-Means data
from sklearn.cluster import KMeans
```

```python
#K=3, red, green and blue clusters
colors = ['red', 'blue', 'green']
kmeans = KMeans(n_clusters=3, random_state=0)
df['cluster'] = kmeans.fit_predict(df[['X', 'Y', 'Z']])
df['c'] = df.cluster.map({0:colors[0], 1:colors[1], 2:colors[2]})
```

```python
#Creating the graph
fig = plt.figure(figsize=(36,10))
ax = fig.add_subplot(131, projection='3d')
ax.scatter(df.X, df.Y, df.Z, c=df.c, s=15)
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
plt.show()
plt.savefig('dataset.png')
```

# Creating a Heatmap of rain since 1901

```python
#Importing the necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Reading the data and assigning the column name
df=pd.read_csv("/content/Rain.csv",header=None)
df.columns=['Date','Actual Avg. Temp','Actual Min. Temp','Actual Max Temp','Average Min.Temp since 1901','Average Max.Temp since 1901','Lowest Temp. since 1901','Highest Temp. since 1901',
            'Actual amount of rain','Average amount of rain since 1901','Highest amount of rain since 1901']
df.head()
```

```python
#Checking for data types and null values
df.info()
```

```python
#Describing the data 
df.describe()
```

```python
#Creeating the correlaltion heatmap
corelation=df.corr()
# Increase the size of the heatmap.
plt.figure(figsize=(16, 6))
# Set the range of values to be displayed on the colormap from -1 to 1
heatmap = sns.heatmap(corelation,vmin=-1, vmax=1, annot=True)
# Give a title to the heatmap. Pad defines the distance of the title from the top of the heatmap.
heatmap.set_title('Correlation Heatmap', fontdict={'fontsize':15}, pad=15);
```
